## Univariate Analysis

```python
for i in df.columns:
    print(i,':',df[i].unique())
```

```python
sns.set_theme(style="darkgrid")
chars=list(df.columns[2:-1])
count={'YES':[],
      'NO':[]}
for i in chars:
    count['YES'].append(round(df[df[i]==2].shape[0]/len(df)*100,2))
    count['NO'].append(round(df[df[i]==1].shape[0]/len(df)*100,2))

width = 0.6  # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(1, 3, figsize=(30, 6), gridspec_kw={'width_ratios': [1, 2.5,1]})
bottom = np.zeros(len(chars))

for yes_no, yes_no_count in count.items():
    p = ax[1].bar(chars, yes_no_count, width, label=yes_no, bottom=bottom)
    bottom += yes_no_count

    ax[1].bar_label(p, label_type='center')

ax[1].set(title='Distribution of Risk Factors and Symptoms',xlabel='Factors and Symptoms', ylabel='Percentage (%)')
ax[1].set_xticks(range(len(chars)))
ax[1].set_xticklabels(chars, rotation=45, ha='right')
ax[1].legend()

age=np.linspace(min(df.AGE)-1,max(df.AGE),11)
df['age group']=pd.cut(df['AGE'], bins=age)
df_age=df['age group'].value_counts().to_frame().reset_index()
df_age['age group']=pd.Categorical(df_age['age group'],ordered=True)
sns.countplot(x=df['age group'], ax = ax[0])
ax[0].set(title='Age distribution')
ax[0].tick_params(axis='x',rotation=45)
ax[0].grid(True)

gender=[len(df[df['GENDER']=='M']),len(df[df['GENDER']=='F'])]
label=['Male','Female']
ax[2].pie(gender,labels=label,autopct='%1.2f%%',startangle=90)
ax[2].set(title='Gender Distribution')

plt.show()
```

```python
import scipy.stats as stats
chars=list(df.columns[2:-1])
expected=[len(df)/2,len(df)/2]
sig_yes =[]
sig_no=[]
no_sig=[]
for i in chars:
    observed=[df[df[i]==2].shape[0],df[df[i]==1].shape[0]]
    chi2_stat,p_value=stats.chisquare(observed, expected)
    
    if p_value<0.05:
        if observed[0]>observed[1]:
            sig_yes.append(f'{i.lower()} (p-value = {p_value:.4f})')
        else:
            sig_no.append(f'{i.lower()} (p-value = {p_value:.4f})')
    else:
        no_sig.append(f'{i.lower()} (p-value = {p_value:.4f})')

chi2_stat,p_value=stats.chisquare(gender, expected)
if p_value<0.05:
    print(f'There is a statistically significant difference (male > female) in the distribution of gender (p-value = {p_value:.4f})\n')
else:
    print(f"There is no statistically significant difference in the distribution of gender ((p-value = {p_value:.4f})\n")
print(f'There is a statistically significant difference (YES > NO) in the distribution of: {sig_yes}\n')
print(f'There is a statistically significant difference (NO > YES) in the distribution of: {sig_no}\n')
print(f"There is no statistically significant difference in the distribution of: {no_sig}")
```

        - This highlights the need for lung cancer awareness and prevention efforts for **both genders**, not just traditionally high-risk groups.


##### 3. Distribution of Risk Factors and Symptoms

#### Men or Women are more likely to get lung-cancer?
(among men surveyed how many percentage got lung cancer?)

```python
men = df[['GENDER','lung_cancer']]
men=men[men['GENDER']=='M'].value_counts().to_frame().reset_index()

women = df[['GENDER','lung_cancer']]
women=women[women['GENDER']=='F'].value_counts().to_frame().reset_index()

fig, ax=plt.subplots(1,2,figsize=(10,6))
ax[0].pie(labels=['YES','NO'],x=men['count'],autopct=lambda p: '{:.2f}%\n({:,.0f})'.format(p, p * sum(men['count']) / 100))
ax[1].pie(labels=['YES','NO'],x=women['count'], autopct=lambda p: '{:.2f}%\n({:,.0f})'.format(p, p * sum(women['count']) / 100))
ax[0].set(title='Percentage of lung-cancer in male')
ax[1].set(title='Percentage of lung-cancer in female');

contingency_table = pd.crosstab(df['GENDER'], df['lung_cancer'])
chi2,p,dof,expect = stats.chi2_contingency(contingency_table)
if p<0.05:
    print(f'There is a statistically significant difference (male > female) in the probability to have lung-cancer (p-value = {p:.4f})\n')
else:
    print(f'There is no statistically significant difference in the probability to have lung-cancer between male and female (p-value = {p:.4f})\n')
```

**Key insights:**
- **89.51%** of males and **85%** of females in the dataset have lung cancer.
- The difference **(~4.5%), p = 0.3122 > 0.05** is **not substantial**, meaning gender alone may not be a strong determining factor for lung cancer in this dataset.
- This suggests that **both men and women are at high risk**, likely due to shared environmental, genetic, or lifestyle factors beyond just gender-based exposure (e.g., smoking rates).

```python
chars=list(df.columns[2:-2]) #species
yes_no_ = {'symptomatic':[],
          'asymptomatic':[]}
for i in chars:
    yes_no_['symptomatic'].append(round(len(df[(df[i]==2) & (df['lung_cancer']==2)])/len(df[df[i]==2])*100,2))
    yes_no_['asymptomatic'].append(round(len(df[(df[i]==1)& (df['lung_cancer']==2)])/len(df[df[i]==1])*100,2))

x = np.arange(len(chars))  # the label locations
width = 0.3  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(figsize=(18,6))

for attribute, measurement in yes_no_.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage of lung-cancer (%)')
ax.set_title('Percentage of people with lung cancer among those who have/do not have the symptoms.')
ax.set_xticks(x + width, chars)
ax.legend(loc='upper left', ncols=3)
ax.tick_params(axis='x',rotation=45)
ax.set_ylim(60, 100)

plt.show()
```

```python
#statistical test
sig_=[]
no_sig_=[]
for i in chars:    
    contingency_table = pd.crosstab(df[i], df['lung_cancer'])
    chi2,p,dof,expect = stats.chi2_contingency(contingency_table)
    if p<0.05:
        sig_.append(f'{i.lower()} (p-value = {p:.4f})')
    else:
        no_sig_.append(f'{i.lower()} (p-value = {p:.4f})')

print(f'There is a statistically significant difference (symptomatic > asymptomatic) in the probability to have lung-cancer in patients with those symptoms: {sig_}\n')
print(f'There is no statistically significant difference in the probability to have lung-cancer between symptomatic and asymptomatic patients with those symptoms: {no_sig_} \n')

```

## Multivariate
#### Hypothesis: peer-pressure lung cancer risk in some age group

```python
new = df[['age group','PEER_PRESSURE','lung_cancer']]

value = {'have peer-pressure':[],
       'not have peer-pressure':[]}
label = new['age group'].unique().sort_values()
for i in label:
    new1 = new[new['age group']==i]
    if len(new1[new1['PEER_PRESSURE']==2]) !=0:
        value['have peer-pressure'].append(len(new1[(new1['PEER_PRESSURE']==2)&(new1['lung_cancer']==2)])/len(new1[new1['PEER_PRESSURE']==2])*100)
        value['not have peer-pressure'].append(len(new1[(new1['PEER_PRESSURE']==1)&(new1['lung_cancer']==2)])/len(new1[new1['PEER_PRESSURE']==1])*100)
    else:
        value['have peer-pressure'].append(0)
        value['not have peer-pressure'].append(0)

x = np.arange(len(label))  # the label locations
width = 0.3  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(figsize=(12,6))

for attribute, measurement in value.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage of lung-cancer (%)')
ax.set_xlabel('Age group')
ax.set_title('The lung-cancer incidence rate with/without peer pressure by age group')
ax.set_xticks(x + width, label)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 120)

plt.show()
```

##### **Key Observations**
- **Youngest group (20.0 - 26.7):** No lung cancer cases → no conclusion.
- **Middle-aged groups (53.5 - 66.9):** Noticeable difference (~10-15%) in lung cancer rates between peer pressure vs. no peer pressure.
- **Older groups (73.6-80):** 100% lung cancer in both cases → likely due to age rather than peer pressure.

### Conclusion
- Peer pressure does not show a statistically significant association with lung cancer risk in any age group analyzed (least p-values: 0.9018, 0.1006, 0.2410, 0.7199).

- Age alone does not explain lung cancer prevalence differences since there is no significant difference in age distribution between the groups.

- Future research should investigate other potential confounding variables (e.g., smoking habits, genetic predisposition, environmental exposure) to gain a deeper understanding of lung cancer risk factors.
