## Multivariate
#### Hypothesis: Peer pressure increases lung cancer risk in some age group

```python
df_age_ = df[['age group',
             'PEER_PRESSURE',
             'lung_cancer']]
```

```python
df_age_ = df_age_.copy()
df_age_["category"] = df_age_[["PEER_PRESSURE", "lung_cancer"]].apply(tuple, axis=1)
df_age_["category"] = df_age_["category"].astype("category")
dg=df_age_.groupby('age group', observed=True)['category'].value_counts().to_frame()

df_unstacked = dg['count'].unstack(level=1)  # Chuyn categories thnh ct
df_unstacked['lung cancer rate in each age group having peer pressure (%)'] = round(df_unstacked[(2,1)]*100 / (df_unstacked[(2,1)] + df_unstacked[(2,2)]),2)
df_unstacked['lung cancer rate in each age group not having peer pressure (%)'] = round(df_unstacked[(1,1)]*100 / (df_unstacked[(1,1)] + df_unstacked[(1,2)]),2)
df_unstacked[['lung cancer rate in each age group having peer pressure (%)','lung cancer rate in each age group not having peer pressure (%)']]
```

#### **Analysis of Lung Cancer Rate by Age Group and Peer Pressure**

##### **Key Observations**
1. **In younger age groups (below 54 years old),** lung cancer rates are either **0%** or significantly **higher in individuals without peer pressure**.
2. **In the 54-73.8 age range,** individuals **without peer pressure consistently have a higher lung cancer rate** compared to those with peer pressure.
3. **Unexpectedly, in the 40.8-47.4 and 80.4-87.0 age groups,** lung cancer is only observed in individuals **without peer pressure (100%)**, raising questions about other influencing factors.

##### **Possible Explanations**
- The results **contradict the hypothesis** that **peer pressure increases lung cancer risk**.
- External factors (e.g., smoking habits, occupational hazards, or lifestyle differences) may play a more significant role in lung cancer risk than peer pressure alone.
- Differences in **age distribution** between the peer pressure and non-peer pressure groups may be affecting the observed lung cancer rates.
  
##### **Next Steps**
1. **Check the age distribution** between peer pressure and non-peer pressure groups to determine if older individuals are overrepresented in one group.
2. **Conduct a Chi-square test** to assess whether the observed differences are statistically significant.
3. **Investigate confounding variables** such as smoking frequency, occupational exposure, and family history.
4. **Perform stratified analysis** to compare lung cancer risk within specific age subgroups, controlling for potential biases.

##### Statistical Test: T-test for Age Difference
- **Null Hypothesis (H):** There is no significant difference in age between the two groups.
- **Alternative Hypothesis (H):** There is a significant difference in age between the two groups.
- **P-value:** `0.7435`

##### Interpretation
- The p-value is **0.74**, which is much higher than the conventional significance threshold (0.05).
- This means **we fail to reject the null hypothesis**  **There is no significant difference in age distribution between the two groups**.
- Thus, **age is not a confounding factor** in explaining differences in lung cancer rates.



```python
import statsmodels.api as sm
df["LUNG_CANCER"] = df["lung_cancer"].map({2: 1, 1: 0})

df["PEER_PRESSURE"] = df["PEER_PRESSURE"].astype(int)
df["AGE"] = df["AGE"].astype(float)
# nh ngha bin c lp (X) v bin ph thuc (y)
X = df[["PEER_PRESSURE", "AGE"]]  # Peer Pressure v tui
y = df["LUNG_CANCER"]  # Ung th phi (1: C, 0: Khng)

# Thm intercept vo m hnh
X = sm.add_constant(X)

# Chy m hnh hi quy logistic
model = sm.Logit(y, X)
result = model.fit()

# Hin th kt qu
print(result.summary())
print(np.exp(result.params))  # Tnh odds ratio
```

#### Logistic Regression Analysis Report

##### 1. Introduction
This analysis examines the relationship between lung cancer diagnosis (`LUNG_CANCER`) and two potential influencing factors:
- **PEER_PRESSURE** (Whether an individual experiences peer pressure)
- **AGE** (The age of the individual)

A logistic regression model was used to assess the significance of these factors.

##### 2. Model Summary

| Model Metric      | Value  |
|------------------|--------|
| Number of Observations | 309 |
| Method | Maximum Likelihood Estimation (MLE) |
| Log-Likelihood | -110.42 |
| Pseudo R-squared | 0.05747 |
| LLR p-value | 0.001191 |

##### 3. Coefficients & Statistical Significance

| Variable       | Coefficient | Std. Error | z-score | P-value | 95% Confidence Interval |
|---------------|-------------|------------|---------|---------|-------------------------|
| **Intercept (const)** | -1.7254 | 1.392 | -1.240 | 0.215 | [-4.453, 1.002] |
| **PEER_PRESSURE** | 1.2128 | 0.388 | 3.128 | **0.002** | [0.453, 1.973] |
| **AGE** | 0.0319 | 0.021 | 1.526 | 0.127 | [-0.009, 0.073] |

##### 4. Interpretation of Results

- **PEER_PRESSURE is a significant predictor of lung cancer** (p-value = **0.002** < 0.05).
  - The odds ratio for `PEER_PRESSURE` is **3.36**, meaning individuals experiencing peer pressure are **3.36 times more likely** to have lung cancer compared to those without peer pressure.
- **AGE is not statistically significant** in predicting lung cancer (p-value = **0.127** > 0.05).
  - This suggests that within this dataset, age does not play a major role in lung cancer risk when peer pressure is accounted for.
- **Intercept (const) is not statistically significant** (p-value = **0.215**), though this is less critical for interpretation.
- **Pseudo R-squared = 0.05747**, indicating that the model explains approximately **5.75%** of the variance in lung cancer occurrence.
  - This suggests that there are other important factors influencing lung cancer that are not included in this model.

##### 5. Explanation of the Paradox

The initial expectation was that age would explain the relationship between peer pressure and lung cancer, but:
- The overall lung cancer rate is **higher in the peer pressure group**.
- However, when analyzed within **individual age groups**, the lung cancer rate in the peer pressure group appears lower.
- The logistic regression results confirm that **peer pressure has an independent effect on lung cancer risk**, and age does not significantly contribute to the model.

This suggests that **peer pressure itself may be a direct risk factor for lung cancer**, possibly due to behaviors or lifestyle choices associated with peer influence (e.g., smoking, stress, etc.).

##### 6. Conclusion
- **Peer pressure significantly increases lung cancer risk**, independent of age.
- **Age alone does not explain the observed differences in lung cancer rates**.
- Future studies should investigate other confounding variables (e.g., smoking habits, genetic predisposition, environmental factors) to further understand this relationship.
